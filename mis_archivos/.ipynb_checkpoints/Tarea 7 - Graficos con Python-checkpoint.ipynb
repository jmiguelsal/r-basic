```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
```

```python
n = 256

X = np.linspace(-np.pi, np.pi, n, endpoint=True)

Y = np.sin(2 * X) 
```

```python
plt.figure(figsize=(10,6), dpi = 300)

plt.plot(X, Y + 1, color='blue', alpha=1.00)

plt.plot(X, Y - 1, color='blue', alpha=1.00)

ax = plt.gca()

ax.fill_between(X, 1, Y+1, facecolor="cornflowerblue")

ax.fill_between(X, -1, Y-1, where=((Y-1)>-1), facecolor="cornflowerblue")
ax.fill_between(X, -1, Y-1, where=((Y-1)<-1), facecolor="pink")

plt.show()
```
