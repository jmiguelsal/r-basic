# ManipulaciÃ³n de data frames

```python
df = pd.DataFrame({
    'key' : ['A','B','C','A','B','C'],
    'value1' : range(6),
    'value2' : np.random.randint(0, 100, 6)
})
```

```python
df
```

```python
df.groupby('key').aggregate(['min', np.median, np.mean, max])
```

```python
df.groupby('key').aggregate({'value1':'min', 'value2':'max'})
```

```python
df.groupby('key').aggregate(min)
```

# Filter

```python
def filter_func(x):
    return x['value2'].std()>5
```

```python
display('df', "df.groupby('key').std()", "df.gropupby('key').filter(filter_func)")
```

```python
df.groupby('key').filter(filter_func)
```

# Transform

```python
df.groupby('key').transform(lambda x: x - x.mean())
```

```python
def norm_by_col2(x):
    x['value1'] /= x['value2'].sum()
    return x
```

```python
df.groupby('key').apply(norm_by_col2)
```
